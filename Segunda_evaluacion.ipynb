```python
# importa las librerías a nuestro entorno de trabajo
from google.colab import drive #Libreria para entorno de ejecución en Colab
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

file_path = 'https://raw.githubusercontent.com/DonYosef/BD-diamantes/main/diamonds.csv' #Base de datos cargada en GitHub para trabajarla en conjunto por medio de colab
df = pd.read_csv(file_path)
df.info() #Por medio del df.info() obtenemos el total de variables con las que contamos en nuestra base de datos, mas la cantidad de datos (no nulas) por cada variable:)
```

```python
#Frecuencia de diamantes en función del precio calculado
plt.figure(figsize=(10, 4))
plt.hist(df['price'], bins=33, color='skyblue', edgecolor='black')
plt.title('Histograma del Precio de los Diamantes')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
#Conteo de la variable "cut" por categoria única
cut_data = df['cut']
print(cut_data.value_counts())
print("--------------------")
cut_data.value_counts().plot(kind='bar')
plt.title('Distribución de la variable "cut"')
plt.xlabel('Cut')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Grafico de torta para representar cantidad de diamantes, en sentido del relojos de mejor a peor calidad
plt.figure(figsize=(8, 6))
df['color'].value_counts().plot.pie(autopct='%1.1f%%', startangle=150, colors=sns.color_palette('viridis'))
plt.title('Proporción de Diamantes por Color')
plt.ylabel('')
plt.show()
```

```python
#Datos relevantes de variable "clarity"
#Conteo por categoria, de menor a mayor valor
clarity_counts = df.groupby('clarity').size()
print(clarity_counts)
```
