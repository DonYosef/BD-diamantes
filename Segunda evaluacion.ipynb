```python
# importa las librerías a nuestro entorno de trabajo
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diamonds.csv", sep=",")
df.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(x = df['cut'], bins = 10)
plt.xlabel("El corte de los diamantes")
plt.ylabel("Cantidad de diamantes")
plt.title("Cantidad de diamantes segun su corte")
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.hist(df['price'], bins=30, color='skyblue', edgecolor='black')  # Puedes ajustar el número de bins según tus necesidades
plt.title('Histograma del Precio de los Diamantes')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
