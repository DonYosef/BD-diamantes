```python
# importa las librer√≠as a nuestro entorno de trabajo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diamonds.csv", sep=",")
df.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(x = df['cut'], bins = 10)
plt.xlabel("El corte de los diamantes")
plt.ylabel("Cantidad de diamantes")
plt.title("Cantidad de diamantes segun su corte")
plt.show()
```

### La mayor cantidad de diamantes es con corte ideal y la menor cantidad de diamantes tienen un corte regular.

```python
plt.figure(figsize=(5,5))
corr = df.corr(method ='pearson');
plt.figure(figsize=(10,10))  # tama√±o del gr√°fico
sns.heatmap(data=corr,square=True, annot=True,cmap='coolwarm');
```

```python
#Gr√°fico 3 torta
df['color'].value_counts().plot.pie(figsize=(5,5), autopct="%0.1f %%");
```

```python
plt.figure(figsize=(15,2))
sns.boxplot(x=df['price'], fliersize=5, width=0.5);
```
